```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_df = pd.read_csv('../athletic_sales_analysis/Resources/athletic_sales_2020.csv')
athletic_sales_2021_df = pd.read_csv('../athletic_sales_analysis/Resources/athletic_sales_2021.csv')
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_df
```

#### Check the data types of each DataFrame

```python
print(athletic_sales_2020_df.dtypes)
```

```python
# Check the 2021 sales data types.
print(athletic_sales_2020_df.dtypes)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
athletic_sales_2020_2021_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], axis=0)
athletic_sales_2020_2021_df
```

```python
# Check if any values are null.
check_null_values = athletic_sales_2020_2021_df.isnull().count()
print(check_null_values)
athletic_sales_2020_2021_df.info()
```

```python
# Check the data type of each column
print(athletic_sales_2020_2021_df.dtypes)
```

```python
# Convert the "invoice_date" to a datetime datatype
athletic_sales_2020_2021_df["invoice_date"] = pd.to_datetime(athletic_sales_2020_2021_df["invoice_date"])
```

```python
# Confirm that the "invoice_date" data type has been changed.
print(athletic_sales_2020_2021_df["invoice_date"].dtypes)
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
Total_Products_Sold = athletic_sales_2020_2021_df.groupby(["region", "state", "city"]).agg(Total_Products_Sold=("units_sold", 'sum')).reset_index()


# Show the top 5 results.
Total_Products_Sold.head(5)                         
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
Number_Products = pd.pivot_table(athletic_sales_2020_2021_df, 
                                    index=["region", "state", "city"], 
                                    values="product",
                                    aggfunc="sum").reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
athletic_sales_2020_2021_df.rename(index={"units": "Total_Products_Sold"})

# Show the top 5 results.
Number_Products.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
Total_Sales = athletic_sales_2020_2021_df.groupby(["region", "state", "city"]).agg(Total_Sales=("total_sales", 'sum')).reset_index()
athletic_sales_2020_2021_df.rename(index={"total_sales": "Total Sales"})

# Show the top 5 results.
Total_Sales.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
Total_Sales_PT = pd.pivot_table(athletic_sales_2020_2021_df, 
                                                    index=["region", "state", "city"], 
                                                    values="total_sales",
                                                    aggfunc="sum").reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.
Total_Sales_PT.head(5)
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
Total_Sales_Retail = athletic_sales_2020_2021_df.groupby(["retailer", "region", "state", "city"]).agg(Total_Sales=("total_sales", 'sum')).reset_index()
athletic_sales_2020_2021_df.rename(index={"total_sales": "Total Sales"})
# Show the top 5 results.
Total_Sales_Retail.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
Total_Sales_Retail_PT = pd.pivot_table(athletic_sales_2020_2021_df, 
                             index=["retailer", "region", "state", "city"],
                             values="total_sales",
                             aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.
Total_Sales_Retail_PT.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
athletic_sales_2020_2021_shoes_df = athletic_sales_2020_2021_df.loc[athletic_sales_2020_2021_df["product"] == "Women's Athletic Footwear"]
athletic_sales_2020_2021_shoes_df.info()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_athletic_shoes_sales = athletic_sales_2020_2021_shoes_df.groupby(["product", "retailer", "region", "state", "city"]).agg(Total_Sales=("total_sales", 'sum')).reset_index()
athletic_sales_2020_2021_shoes_df.rename(index={"units_sold": "Womens_Footwear_Units_Sold"})

# Show the top 5 results.
womens_athletic_shoes_sales.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_athletic_shoes_sales_PT = pd.pivot_table(athletic_sales_2020_2021_shoes_df, 
                             index=["product", "retailer", "region", "state", "city"],
                             values="total_sales",
                             aggfunc='sum').reset_index()
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
athletic_sales_2020_2021_shoes_df.rename(index={"units_sold": "Womens_Footwear_Units_Sold"})

# Show the top 5 results.
womens_athletic_shoes_sales_PT.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_athletic_shoes_day_sales_PT = pd.pivot_table(athletic_sales_2020_2021_shoes_df, 
                             index="invoice_date",
                             values="total_sales",
                             aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.
womens_athletic_shoes_day_sales_PT.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
womens_athletic_shoes_daily_sales = womens_athletic_shoes_day_sales_PT.resample("D", on="invoice_date").sum()
# Sort the resampled pivot table in ascending order on "Total Sales".
womens_athletic_shoes_daily_sales_sorted = womens_athletic_shoes_daily_sales.sort_values("total_sales")
womens_athletic_shoes_daily_sales
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
womens_athletic_shoes_weekly_sales = womens_athletic_shoes_day_sales_PT.resample("W", on="invoice_date").sum()
womens_athletic_shoes_weekly_sales

# Sort the resampled pivot table in ascending order on "Total Sales".
womens_athletic_shoes_weekly_sales_sorted = womens_athletic_shoes_weekly_sales.sort_values("total_sales")
womens_athletic_shoes_weekly_sales
```
